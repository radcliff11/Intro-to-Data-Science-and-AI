```python
import numpy as np
import pandas as pd
```

```python
#simi for similarities
```

```python
simi = pd.read_excel("MeanAbsoluteDifference among countries.xls")
```

```python
simi = simi.replace(0,10)
```

```python
cols = simi.columns
rows = simi.index
```

```python
def find_similar(eps):
    pair = []
    for i in range(len(cols)):
        for k in range(len(rows)):
            if simi.iloc[k,i] < eps and abs(k-i) >20:
                pair.append((cols[i],rows[k],i,k))
    return pair
```

```python
find_similar(0.09) #find the countries pair that are very similar accross the 6 main predictors
```
