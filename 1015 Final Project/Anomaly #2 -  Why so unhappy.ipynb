```python
non_nan_cols =[]
for i in Happy2018.columns:
    if not Happy2018[i].isnull().any():
        non_nan_cols.append(i)
```

```python
big_6 = [ 'Log GDP per capita', 
             'Social support', 
         'Healthy life expectancy at birth', 
         'Freedom to make life choices', 
         'Generosity',
         'gini of household income reported in Gallup, by wp5-year']
```

```python
for i in big_6:
    print("This is stats for ",i)
    a = pd.DataFrame(Happy2018[i])
    print(a.describe())
    print()
```

```python
big6_std = [10.282658,0.905436,69.051659,0.883905,0.095793,0.552189 ] #The upper quartile of each of the 6 predictors
```

```python
def why_so_unhappy(eps,rank):
    unhappy = []
    countries_names = list(Happy2018["country"])
    for i in range(len(countries_names)):
        country = Happy2018.iloc[i]
        diff = [(x-y)/y for x,y in zip(country[big_6], big6_std) ]
        result = sum(diff)/len(diff)
        if result > eps and i > rank:
            unhappy.append(countries_names[i])
    return unhappy
        
```

```python
unhappy_countries = why_so_unhappy(0,70) #0: means they are appx better than 75% of the rest in terms of big 6 , i.e they are > Q1 
                    #70 means they score rank 50 and below
```
